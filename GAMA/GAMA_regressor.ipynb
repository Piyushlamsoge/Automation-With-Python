```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from gama import GamaRegressor
from sklearn.metrics import mean_squared_error
```

```python
X , y = load_diabetes(return_X_y=True , as_frame = True)
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(X , y , random_state = 0)
```

```python
custom_automl = GamaRegressor(max_total_time=180 , n_jobs = 1)
```

```python
custom_automl.fit(xtrain , ytrain)
```

```python
prediction =custom_automl.predict(xtest)
```

```python
print("Mean_Squared_Error:", mean_squared_error(ytest, prediction))
```
