```python
import pandas as pd 
import numpy as np
train_data = pd.read_excel('/content/Data_Train.xlsx')
train_data.head()
```

```python
from AutoClean import AutoClean
clean_1 = AutoClean(train_data)
```

```python
train_data['Reviews']= pd.to_numeric(train_data['Reviews'].apply(lambda x: x[:-len('out of 5 stars')]))
```

```python
train_data
```

```python
train_data['Ratings'] = pd.to_numeric(train_data['Ratings'].apply(lambda y : y.replace(",","")[:-len('customer reviews')]))
```

```python
train_data
```

```python
from autogluon.multimodal import MultiModalPredictor
predictor = MultiModalPredictor(label = 'Price')
predictor.fit(train_data, time_limit=3 * 60)
```
