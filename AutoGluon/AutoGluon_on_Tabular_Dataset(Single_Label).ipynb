```python
from tables import index
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
#train_data = TabularDataset('https://raw.githubusercontent.com/jbrownlee/Datasets/master/adult-train.csv' )

train_data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/adult-train.csv' , header = None )
#column = [ 'age' , 'workclass' , 'fnlwgt' , 'education', 'education_num' , 'marital_status' , 'occupation' , 'relationship' , 'race' , 'sex' , 'capital_gain' , 'capital_loss' , 'hours_per_week' , 'native_country' , 'class']

#train_data = pd.DataFrame(train_data , columns= [ 'age' , 'workclass' , 'fnlwgt' , 'education', 'education_num' , 'marital_status' , 'occupation' , 'relationship' , 'race' , 'sex' , 'capital_gain' , 'capital_loss' , 'hours_per_week' , 'native_country' , 'class'])
train_data.head()
```

```python
label = 14
train_data[label].describe()
```

```python
predictor = TabularPredictor(label = label).fit(train_data)
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/adult-test.csv' , header = None)
y_test = test_data[label]
x_test = test_data.drop(columns = [label])
x_test.head()
```

```python
y_pred = predictor.predict(x_test)
#result = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True , average='micro' )
```
