```python
import pandas as pd
df = pd.read_csv('/content/world-cities_csv.csv')
df.head()
```

```python
from pycaret.clustering import *
s = setup(df , normalize = True ,  high_cardinality_features = ['name' , 'country' , 'subcountry']  ) 
```

```python
model = create_model('kmeans')
```

```python
evaluate_model(model)
```

```python
assign_model(model)
```

```python
predict_model(model, data = df)
```
