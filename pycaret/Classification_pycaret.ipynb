```python
from pydataset import data
data()
```

```python
df = data('iris')
df.head()
```

```python
from pycaret.classification import *
s = setup(df , target ='Species' , numeric_features = ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'] , normalize = True ,remove_outliers= True , pca = True , feature_interaction = True , feature_ratio = True , fold_strategy= 'kfold' , fold_shuffle = True)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```
